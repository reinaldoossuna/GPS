```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
plt.style.use("bmh")
```

```python
grace = pd.read_csv("DATA/GRACE/STATIONS/BRAZ.csv",index_col="DATE",parse_dates=["DATE"]) 
gnss = pd.read_csv("DATA/gnss_braz.csv",index_col="DATE",parse_dates=["DATE"])
```

```python
fig, ax1 = plt.subplots() 
color = 'tab:red' 
ax1.plot(grace.index,grace.mean(axis=1),color=color) 
ax1.tick_params(axis='y', labelcolor=color) 

ax2 = ax1.twinx() 
color = 'tab:blue' 
ax2.plot(gnss.index, gnss.Altitude_smooth) 
ax2.tick_params(axis='y', labelcolor=color);
```
